### Step:04
### ● Visualize the dataset:

○ Create a scatter plot to show relationships between features.

```python
sns.scatterplot(x='sepal_length' , y='sepal_width' , hue='species' , data=df)
plt.grid()
plt.show()

print('Difference B/W')

sns.scatterplot(x='petal_length' , y='petal_width' , hue='species' , data=df)
plt.grid()
plt.show()

```

○ Use histograms to show value distributions.

```python
# Show histrogram from method:01
sns.histplot(data=df , bins=16)
plt.show()
```

Difference B/W

```python
# Show histrogram from method:02
df.hist()
plt.show()
```

○ Use box plots to identify outliers.

```python
# Find Outliers 
sns.boxplot(df)
plt.grid()
plt.show()
print('sepal_width feature = outliers')
```
